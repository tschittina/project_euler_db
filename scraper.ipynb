```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# Initialize lists to store data
languages = []
lang_n_members = []
lang_mean_percent_solved = []

cols = [languages, lang_n_members, lang_mean_percent_solved]

# Tags for langauge table
tags = ['td.language_column', 'td.users_column', 'td.mean_column']
```

```python
# PHPSESSID from current session is saved in .txt file
# Need to update to for every new session
with open('sessionID.txt', 'r') as f:
    cookies = f.read().strip()

# Create session object and establish cookies
session = requests.Session()
cookies_dict = {'PHPSESSID': cookies}

# Request language page
url = 'https://projecteuler.net/languages'
lang_response = session.get(url, cookies=cookies_dict)
if lang_response.status_code != 200:
    raise Exception(f"Failed to retrieve language page: {lang_response.status_code}")

# Parse the HTML content
soup = BeautifulSoup(lang_response.content, 'html.parser')
for col, tag in zip(cols, tags):
    for item in soup.select(tag):
        col.append(item.string.strip())
```

```python
# Create a DataFrame with language data
lang_data = pd.DataFrame({
    'langauge': languages,
    'n_members': lang_n_members,
    'mean_percent_solved': lang_mean_percent_solved
})

lang_data
```

```python
# Repeat process for location data
locations = []
loc_n_members = []
loc_mean_percent_solved = []

cols = [locations, loc_n_members, loc_mean_percent_solved]

tags = ['td.location_column', 'td.user_column', 'td.mean_column']


url = 'https://projecteuler.net/locations'
loc_response = session.get(url, cookies=cookies_dict)
if loc_response.status_code != 200:
    raise Exception(f"Failed to retrieve location page: {loc_response.status_code}")

soup = BeautifulSoup(loc_response.content, 'html.parser')
for col, tag in zip(cols, tags):
    for item in soup.select(tag):
        col.append(item.text.strip())

loc_data = pd.DataFrame({
    'location': locations,
    'n_members': loc_n_members,
    'mean_percent_solved': loc_mean_percent_solved
})

loc_data
```

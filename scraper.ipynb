```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# Initialize lists to store data
languages = []
lang_n_members = []
mean_percent_solved = []

cols = [languages, lang_n_members, mean_percent_solved]

# Tags for langauge table
tags = ['td.language_column', 'td.users_column', 'td.mean_column']
```

```python
# PHPSESSID from current session is saved in .txt file
# Need to update to for every new session
with open('sessionID.txt', 'r') as f:
    cookies = f.read().strip()

# Create session object and establish cookies
session = requests.Session()
cookies_dict = {'PHPSESSID': cookies}

# Request language page
url = 'https://projecteuler.net/languages'
response = session.get(url, cookies=cookies_dict)
if response.status_code != 200:
    raise Exception(f"Failed to retrieve language page: {response.status_code}")

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
for col, tag in zip(cols, tags):
    for item in soup.select(tag):
        col.append(item.string.strip())
```

```python
# Create a DataFrame with language data
lang_data = pd.DataFrame({
    'langauge': languages,
    'n_members': lang_n_members,
    'mean_percent_solved': mean_percent_solved
})

lang_data
```
